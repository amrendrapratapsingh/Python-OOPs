## Inheritance

    -An Application may contains multiple classes and classes may contain some common features, if there are some common  features ,then instead of duplicating the common features in each class,we can extract the common features into a super class and extend it to other sub-class, this process called `inheritence`

#### Why to inherit

    -To Avoid Redundancy.
    -To Get Reuseability.

Syntax for creating a subclass:
    
    class subclass(superclass):
            -
            -
            (or)
     class subclass(superclass1, superclass2, ..,n)
     

 1) In Python we can create object for superclass or subclass.
 2) In Python we can't store subclass object in superclass references or we can't Typrecast supercast references to subclass reference,because in Python we don't have upcasting and downcasting.
 3)By default,all attributes of superclass are inherited to subclass,if we don't want to inherit any attribute then we should make them private one by prefixing '__'

```python
class A:
    def __init__(self,x):
        self.__x=x          #self.x =>private variable
    def display(self):
        print('display')

class B(A):
    def show(self):
        print('Show')
        print('x value = ',self.x) #error ,as private variable can't be inherited
#outside
b = B(100)
b.display()
b.show()
```

    1) In Python ,if we create a subclass object ,then Python will search for constructor in subclass ,if not exist then Ptyhton calls superclass constructor.

```python
class A:
    def __init__(self,x):
        self.x=x          #self.x =>instance variable
    def display(self):
        print('display')

class B(A):
    def show(self):
        print('Show')
        print('x value = ',self.x)
#outside
b = B(100)
b.display()
b.show()
```

 1) If both,superclass and subclass are containing constructors ,then Python always call subclass constructor,whenever subclass object is created.
 2)We can call superclass constructor from subclass constructor with keyword "super()"

```python
class A:
    def __init__(self,x):  #superclass constructor/initializor
        self.x=x          
    def display(self):
        print('display')

class B(A):
    def show(self,x,y):
        self.m = y
        super().__init__(x)  #calling superclass constructor
    def show(self):
        print('Show')
        print('x value = ',self.x)
#outside
b = B(100,200)
b.display()
b.show()
```

#### super() ,can be used to call superclass variables or methods or constructor

```python
class A:         
    def display(self):
        print('display in A')

class B(A):
    def display(self):
        print('display in B')
    def show(self):
        print('Show')
        self.display() #calling ,display of B
        super().display() # calling display of A
#outside
b = B()
b.show()
```

    super(), indicates superclass object, by using this we can call @classmethods and @staticmethod also.

```python
class A:  
    @staticmethod
    def display(self):
        print('display in A')

class B(A):
    def show(self):
        print('Show')
        super().display() # calling display of A
```
